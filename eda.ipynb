```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
women_df = pd.read_csv("Womens Clothing E-Commerce Reviews.csv")
women_df.head()
```

```python
women_df.fillna("..", inplace=True)
women_df

```

```python
plt.figure(figsize=(8,5))
sns.countplot(data = women_df, x="Rating", palette="viridis")
plt.title("Distribution of Ratings")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=women_df, x="Recommended IND", palette="pastel")
plt.title("Recommendation Distribution")
plt.show()
```

```python
women_df[["Age", "Rating"]].head()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data = women_df, x="Rating", y="Age", palette="coolwarm")
plt.title("Age vs Rating")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.violinplot(data=women_df, x='Rating', y='Age', palette='muted')
plt.title("Age vs Rating (Violin Plot)")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(data=women_df, x='Rating', y='Age', palette='coolwarm', estimator=np.mean)
plt.title("Average Age per Rating (Bar Plot)")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data = women_df, x="Recommended IND", hue="Rating", palette="Set1")
```

```python
pd.crosstab(women_df['Recommended IND'], women_df['Rating']) #--> Cross tabulation between Recommended IND and Rating
```

```python

plt.figure(figsize=(8,6))  
numeric_women_df = women_df.select_dtypes(include=['int64', 'float64'])  
sns.heatmap(data=numeric_women_df.corr(), annot=True, cmap='coolwarm')  
plt.title("Correlation Heatmap")  
plt.show()
```
